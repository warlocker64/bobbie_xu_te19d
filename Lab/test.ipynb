```python
def yes (a,b):
    if a < b :
        return a 
    else:
        return b
    
print(yes(10,2))
```

```python
def medelvärde(yes,yes2):
    medel = (yes+yes2)/2 #local variabel.
    return medel

no = medelvärde(70,73)


print(f"yes {no}")


```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2
print(f(5))


x=np.linspace(-10,10)
y=f(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) = x^2")
plt.grid()
plt.show()
```
