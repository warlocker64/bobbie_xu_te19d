```python
#Uppgift 2 abc
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Syssel.csv", header = 1)
män = sum(df.query('sysselsättning == "förvärvsarbetande" and kön == "män"')['2019'])
kvinnor = sum(df.query('sysselsättning == "förvärvsarbetande" and kön == "kvinnor"')['2019'])
intekvinnor = sum(df.query('sysselsättning == "ej förvärvsarbetande" and kön == "kvinnor"')['2019'])
intemän = sum(df.query('sysselsättning == "ej förvärvsarbetande" and kön == "män"')['2019'])
print(f" Antal män: {män}")
print(f" Antal kvinnor: {kvinnor}")
print(f" Antal kvinnor som inte: {intekvinnor}")
print(f" Antal män som inte: {intemän}")

height = [män,kvinnor,intekvinnor,intemän]
bars = ["män","kvinnor","män som inte","kvinnor som inte"]
x_pos = np.arange(len(bars))

plt.bar(x_pos,height,width=.7, color = ["black","pink"])
plt.xticks(x_pos, bars , color='black')
plt.yticks(color='black')

plt.show()

```
